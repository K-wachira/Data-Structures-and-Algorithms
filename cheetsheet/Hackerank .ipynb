```python
# character reprograming 
def getMaxDelections(s):
    directions =[]
    for i in s:
        directions.append(i)
        
    print(len(directions))
    
    u =0
    d =0
    r =0
    l =0
    
    for i in directions:
        if i is "U":
            u+=1
        if i is "D":
            d+=1 
        if i is "R":
            r+=1
        if i is "L":
            l+=1
            
    
    
    newu =0
    newd =0
    newl =0
    newr=0 
    
    if u>d:
      newu =u-d 
    if d>u:
      newd= d-u 
    if r>l:
        newr =r-l
    if l>r:
        newl= l-r
        
    total = newu+newd+newr+newl
    
    return (len(directions)-total)
        
        
    
    
s = ("UDRLDLUDRL")    
getMaxDelections(s)
    
```

```python
# Dublicated products

def numDuplicates(name, price, weight):
    
#     create a dict 
# if value of dict is more than one 

#     check if price and weight are same for more than one 
#     if yes reduce the value 
#     increase the counter to be returned

    dict1 =list(zip(name,price,weight))
    dict2 = set(dict1)
    return len(dict1)-len(dict2)


    
    
name = ["ball", "bat", "glove", "glove", "glove"]
price = [2, 3, 1, 2, 1]
weight = [2, 5, 1, 1, 1]    
numDuplicates(name, price, weight)
```

```python
def string_match(a, b):
  shorter = min(len(a), len(b))
  count = 0
  
  # Loop i over every substring starting spot.
  # Use length-1 here, so can use char str[i+1] in the loop
  for i in range(shorter-1):
    a_sub = a[i:i+2]
    b_sub = b[i:i+2]
    if a_sub == b_sub:
      count = count + 1

  return count

a ="abc"
b ="abc"
string_match(a,b)
    
            
```
