# Merging Two sorted Linked List into One sorted Linked List

```python
# Definition for singly-linked list.
class ListNode(object):
     def __init__(self, val=0, next=None):
         self.val = val
         self.next = next
            
class Solution(object):
    def reverseList(self, l1,l2):
        newhead  = ListNode(0) 
        node = newhead
        
        while l1 and l2: 
            if l1.val > l2.val:
                node.next = l2
                node = node.next
                l2 = l2.next
            else:                
                node.next = l1
                node = node.next
                l1 = l1.next 
        while l1:
            node.next = l1
            node = node.next
            l1 = l1.next
        while l2:
            node.next = l2
            node = node.next
            l2 = l2.next
        node.next = None
        newhead = newhead.next
        
        while newhead:            
            print(str(newhead.val) + " -> ", end = "")
            newhead = newhead.next
head=  ListNode(1)
second = ListNode(4)
third = ListNode(5)
fou = ListNode(6)
fifth = ListNode(7)
head.next = second
second.next = third
third.next = fou
fou.next = fifth
fifth.next = None


head1=  ListNode(1)
second1 = ListNode(2)
third1 = ListNode(3)
fou1= ListNode(8)
fifth1 = ListNode(9)
head1.next = second1
second1.next = third1
third1.next = fou1
fou1.next = fifth1
fifth1.next = None























l1 = head
l2 = head1

while head:
    print(str(head.val) + " -> ", end = "")
    head = head.next
print("\n")   
while head1:
    print(str(head1.val) + " -> ", end = "")
    head1 = head1.next
    
    
print("\n\nresults =\n")
    

obj = Solution()
obj.reverseList(l1, l2)
```
