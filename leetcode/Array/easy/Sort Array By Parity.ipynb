# Sort Array By Parity

[Question Link](https://leetcode.com/problems/sort-array-by-parity/)


Given an array A of non-negative integers, return an array consisting of all the even elements of A, followed by all the odd elements of A.

You may return any answer array that satisfies this condition.
***
## Approach 1(iterate)

# Steps ::
1. loop though the given array and if an element is even append it to the even list and if its odd append it to the odd list
2. when done with the loop append odd list to even list  and return the even list

```python
class Solution(object):
    def sortArrayByParity(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        odd = []
        even = []
        
        for i in A:
            if i% 2 == 0:
                even.append(i)
            elif i % 2 != 0 :
                odd.append(i)
        even.extend(odd)
        return even
        
obj = Solution()
A = [3,1,2,4]

obj.sortArrayByParity(A)
```
